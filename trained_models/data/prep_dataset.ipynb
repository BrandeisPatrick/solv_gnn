```python
import pandas as pd 
```

```python
df = pd.read_csv('/home/dyerlab/Documents/solv_gnn/trained_models/data/src/ChemFluor2024-01-09.csv')
```

```python
# chaneg Chromophore, solvent to SoluteSMILES,SolventSMILES,DeltaGsolv
df = df.rename(columns={'Chromophore': 'SoluteSMILES', 'Solvent': 'SolventSMILES'})
```

```python
df
```

```python
# convert nm to eV 
df['emi'] = 1239.8/df['emi']
df['abs'] = 1239.8/df['abs']

df['DeltaGsolv'] = df['abs'] - df['emi'] 
```

```python
df1 = df[['SoluteSMILES', 'SolventSMILES', 'DeltaGsolv']]
df1.to_csv('chemfluor.csv', index=False)
```
